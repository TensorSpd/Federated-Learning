## Introduction
This notebook demonstrates a Federated Learning simulation using IID and Non-IID data distributions. Federated Learning enables collaborative training of machine learning models across multiple devices or clients while ensuring data privacy. The focus is to analyze the performance of Federated Learning under IID (independent and identically distributed) and Non-IID (non-independent and non-identically distributed) scenarios.

---


## Setup and Imports

```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, losses, metrics
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
from copy import deepcopy
from IPython.display import clear_output
import warnings
import math

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')
tf.get_logger().setLevel('ERROR')

sns.set(style="whitegrid")
%matplotlib inline
```

## Constants and Configuration
Here, we define global constants and configuration parameters, such as:
- Number of clients
- Learning rate
- Number of communication rounds
- Batch size
These settings control the overall behavior of the Federated Learning simulation.

---

```python
# Simulation Parameters 
NUM_CLIENTS = 5
FL_ROUNDS = 10
EPOCHS = 5
BATCH_SIZE = 32
DATA_DISTRIBUTION_TYPES = ['iid', 'non-iid'] 
DROPOUT_PROBABILITY = 0.0 

SEED = 42
np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)
```

## Data Loading and Distribution
This section focuses on:
1. Loading datasets 
2. Preparing data partitions for clients.
3. Implementing IID (independent and identically distributed) and Non-IID (non-independent and non-identically distributed) data distributions.

---

```python
def load_mnist_data():
    (x_train, y_train), (x_test, y_test) = mnist.load_data()
    # Normalize the data to [0, 1]
    x_train = x_train.astype('float32') / 255.0
    x_test  = x_test.astype('float32') / 255.0
    # Flatten the images to 784-dimensional vectors
    x_train = x_train.reshape(-1, 784)
    x_test  = x_test.reshape(-1, 784)
    return (x_train, y_train), (x_test, y_test)

def distribute_data(x, y, num_clients, distribution='iid'):
    data_per_client = int(len(x) / num_clients)
    client_data = {}
    
    if distribution == 'iid':
        # Shuffle the data
        indices = np.arange(len(x))
        np.random.shuffle(indices)
        x, y = x[indices], y[indices]
        for i in range(num_clients):
            start = i * data_per_client
            end = (i + 1) * data_per_client
            client_data[i] = {'x': x[start:end], 'y': y[start:end]}
    elif distribution == 'non-iid':
        # Sort the data by label to create non-iid partitions
        sorted_indices = np.argsort(y)
        x, y = x[sorted_indices], y[sorted_indices]
        shards_per_client = 2  # Number of shards per client
        num_shards = num_clients * shards_per_client
        shard_size = int(len(x) / num_shards)
        for i in range(num_clients):
            client_shards_x = []
            client_shards_y = []
            for _ in range(shards_per_client):
                start = (i * shards_per_client + _) * shard_size
                end = (i * shards_per_client + _ + 1) * shard_size
                client_shards_x.append(x[start:end])
                client_shards_y.append(y[start:end])
            client_data[i] = {'x': np.concatenate(client_shards_x),
                              'y': np.concatenate(client_shards_y)}
    else:
        raise ValueError("Unsupported data distribution type.")
    
    return client_data

# Load data
(x_train, y_train), (x_test, y_test) = load_mnist_data()

# Aggregate global test data
global_test_data = {'x': x_test, 'y': y_test}

# Display data distribution information
for distribution in DATA_DISTRIBUTION_TYPES:
    print(f"\nData Distribution: {distribution.upper()}")
    clients_data = distribute_data(x_train, y_train, NUM_CLIENTS, distribution=distribution)
    for client_id, data in clients_data.items():
        print(f"Client {client_id + 1}: {data['x'].shape[0]} samples")
```

## Utility Functions

```python
def simulate_dropout(probability):
    return random.uniform(0, 100) < probability
```

## Model Definition
We define the architecture of the neural network used in the Federated Learning process which includes a simple model:

- A fully connected neural network (FCNN)

---

```python
def create_model():

    model = models.Sequential([
        layers.Dense(64, activation='relu', input_shape=(784,)),
        layers.Dense(10, activation='softmax')
    ])
    
    model.compile(optimizer=optimizers.Adam(),
                  loss=losses.SparseCategoricalCrossentropy(),
                  metrics=['accuracy'])
    return model
```

## Client Class Definition
The `Client` class encapsulates client-side operations, including:
- Local training on the client's data.
- Maintaining client-specific data partitions.
- Returning updates to the global server.

---

```python
class Client:
    def __init__(self, client_id, data):
        # Initialize a client object with necessary attributes
        self.client_id = client_id
        self.x = data['x']
        self.y = data['y']
        self.model = create_model()
        self.accuracy_history = []
        self.loss_history = []
        self.learning_rate = 0.001  
        self.epochs = EPOCHS  
    
    def set_weights(self, weights):
        # Set model weights for the client
        self.model.set_weights(weights)
    
    def get_weights(self):
        # Retrieve the current model weights from the client
        return self.model.get_weights()
    
    def compile_model(self, learning_rate=0.001):
        # Compile the model with a specified learning rate, loss function, and metrics
        self.model.compile(optimizer=optimizers.Adam(learning_rate=learning_rate),
                           loss=losses.SparseCategoricalCrossentropy(),
                           metrics=['accuracy'])
    
    def train(self):
        # Train the model on the client's local data
        self.compile_model(learning_rate=self.learning_rate)
        history = self.model.fit(self.x, self.y,
                                 epochs=self.epochs,
                                 batch_size=BATCH_SIZE,
                                 verbose=0,
                                 callbacks=[EpochLogger(self)])
        return history
    
    def evaluate(self, x, y):
        # Evaluate the model on the provided data and return accuracy
        loss, acc = self.model.evaluate(x, y, verbose=0)
        return acc

class EpochLogger(tf.keras.callbacks.Callback):
    def __init__(self, client):
        # Initialize the logger to track epoch-level metrics for a specific client
        super(EpochLogger, self).__init__()
        self.client = client
    
    def on_epoch_end(self, epoch, logs=None):
        # Log the accuracy and loss at the end of each epoch
        acc = logs.get('accuracy')
        loss = logs.get('loss')
        self.client.accuracy_history.append(acc)
        self.client.loss_history.append(loss)
        print(f"Client {self.client.client_id + 1} - Epoch {epoch + 1}: Accuracy = {acc:.4f}, Loss = {loss:.4f}")
```

## Server Aggregation Function
This section defines the server's aggregation mechanism. The server collects updates from all clients and combines them to update the global model. Common strategies include:
- **FedAvg (Federated Averaging)**: A weighted average of client updates.
- Other advanced aggregation methods (if applicable).

---

```python
def aggregate_models(clients):
    client_weights = [client.get_weights() for client in clients]
    
    # Initialize the averaged weights list
    averaged_weights = []
    
    # Iterate through each layer's weights
    for layer_weights in zip(*client_weights):
        # Convert to numpy array for averaging
        averaged_layer = np.mean(np.array(layer_weights), axis=0)
        averaged_weights.append(averaged_layer)
    
    return averaged_weights
```

## Federated Learning Simulation Function
This function implements the entire simulation pipeline, including:
1. Distributing data to clients.
2. Running local training on clients.
3. Aggregating updates at the server.
4. Iterating through multiple communication rounds to improve the global model.

---

```python
def run_federated_learning(data_distribution, fl_rounds=FL_ROUNDS, epochs=EPOCHS, dropout_prob=DROPOUT_PROBABILITY):
    print(f"\n=== Starting Federated Learning Simulation: {data_distribution.upper()} ===")
    
    # Step 1: Distribute data among clients
    current_clients_data = distribute_data(x_train, y_train, NUM_CLIENTS, distribution=data_distribution)
    
    # Display data distribution information
    print(f"Data distribution: {data_distribution}")
    for client_id, data in current_clients_data.items():
        print(f"Client {client_id + 1}: {data['x'].shape[0]} samples")
    
    # Step 2: Initialize clients
    current_clients = []
    for client_id in range(NUM_CLIENTS):
        client = Client(client_id, current_clients_data[client_id])
        current_clients.append(client)
    
    # Step 3: Initialize server model
    server_model = create_model()
    
    # Step 4: Initialize global accuracy history
    current_global_accuracy_history = []
    
    # Step 5: Run Federated Learning rounds
    for round_num in range(1, fl_rounds + 1):
        print(f"\n--- Federated Learning Round {round_num} ---")
        
        participating_clients = []
        # Each client trains on local data
        for client in current_clients:
            # Simulate dropout
            if simulate_dropout(dropout_prob):
                print(f"Client {client.client_id + 1} dropped out during communication.")
                continue
            # Set client weights to server weights
            client.set_weights(server_model.get_weights())
            # Set client learning rate (can be parameterized if needed)
            client.learning_rate = 0.001
            # Train client model
            client.train()
            participating_clients.append(client)
        
        if not participating_clients:
            print("No clients participated in this round.")
            continue
        
        # Aggregate client models using Federated Averaging
        averaged_weights = aggregate_models(participating_clients)
        server_model.set_weights(averaged_weights)
        print("Aggregated client models to update the server model.")
        
        # Evaluate global model
        test_loss, test_acc = server_model.evaluate(global_test_data['x'], global_test_data['y'], verbose=0)
        current_global_accuracy_history.append(test_acc)
        print(f"Global Model Accuracy: {test_acc:.4f}")
    
    # Step 6: Collect clients' accuracy and loss histories
    client_accuracy_histories = {client.client_id +1: client.accuracy_history for client in current_clients}
    client_loss_histories = {client.client_id +1: client.loss_history for client in current_clients}
    
    # Compile results
    results = {
        'data_distribution': data_distribution,
        'client_accuracy_histories': client_accuracy_histories,
        'client_loss_histories': client_loss_histories,
        'global_accuracy_history': current_global_accuracy_history
    }
    
    print(f"\n=== Completed Federated Learning Simulation: {data_distribution.upper()} ===\n")
    
    return results
```

## Running Simulations for IID and Non-IID
We execute the Federated Learning simulation for both IID and Non-IID data distributions. This section highlights:
- Performance differences between the two setups.
- Model accuracy, loss, and convergence behavior over multiple rounds.

---

```python
# Run simulation for IID data distribution
iid_results = run_federated_learning(
    data_distribution='iid',
    fl_rounds=FL_ROUNDS,
    epochs=EPOCHS,
    dropout_prob=DROPOUT_PROBABILITY  
)

# Run simulation for Non-IID data distribution
non_iid_results = run_federated_learning(
    data_distribution='non-iid',
    fl_rounds=FL_ROUNDS,
    epochs=EPOCHS,
    dropout_prob=DROPOUT_PROBABILITY  
)
```

## Enhanced Visualization
- Accuracy and loss curves for IID and Non-IID setups.

```python
# Cell 10: Enhanced Visualization

def plot_federated_learning_results(results, title_suffix=''):
    """
    Plots Federated Learning results with subplots for each client and the global model.
    
    Parameters:
    - results (dict): Contains 'data_distribution', 'client_accuracy_histories', and 'global_accuracy_history'.
    - title_suffix (str): Additional title information (e.g., 'IID' or 'Non-IID').
    """
    import matplotlib.pyplot as plt
    import math
    
    data_distribution = results['data_distribution']
    client_histories = results['client_accuracy_histories']
    global_history = results['global_accuracy_history']
    
    num_clients = len(client_histories)
    cols = 3  
    rows = math.ceil(num_clients / cols) + 1  
    
    # Create subplots with adjusted spacing
    fig, axes = plt.subplots(rows, cols, figsize=(cols * 5, rows * 4))
    axes = axes.flatten()  
    
    fig.suptitle(f'Federated Learning Results ({data_distribution.upper()} Data)', fontsize=18, fontweight='bold', y=1.02)
    client_colors = plt.cm.tab10.colors  
    
    # Plot each client's accuracy history
    for idx, (client_id, accuracy_history) in enumerate(client_histories.items()):
        ax = axes[idx]
        if accuracy_history:
            epochs = range(1, len(accuracy_history) + 1)
            accuracies = accuracy_history
            color = client_colors[idx % len(client_colors)]  
            ax.plot(epochs, accuracies, marker='o', color=color, linestyle='-', linewidth=2, markersize=6, label=f'Client {client_id}')
            ax.set_title(f'Client {client_id} Accuracy', fontsize=14, fontweight='medium')
            ax.set_xlabel('Epoch', fontsize=12)
            ax.set_ylabel('Accuracy', fontsize=12)
            ax.set_ylim(0, 1)
            ax.grid(True, linestyle='--', alpha=0.7)
            ax.legend(loc='lower right', fontsize=10)
        else:
            ax.axis('off')  
    
    # Plot Global Accuracy in the last subplot
    global_ax = axes[num_clients]
    global_ax.plot(range(1, len(global_history) + 1), 
                  global_history, 
                  marker='o', color='green', linestyle='-', linewidth=2, markersize=6, label='Global Model')
    global_ax.set_title('Global Model Accuracy', fontsize=14, fontweight='medium')
    global_ax.set_xlabel('FL Round', fontsize=12)
    global_ax.set_ylabel('Accuracy', fontsize=12)
    global_ax.set_ylim(0, 1)
    global_ax.grid(True, linestyle='--', alpha=0.7)
    global_ax.legend(loc='lower right', fontsize=10)
    
    for ax in axes[num_clients + 1:]:
        ax.axis('off')
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    
    # Display the plots
    plt.show()

# Plot results for IID data distribution
plot_federated_learning_results(iid_results, title_suffix='IID')

# Plot results for Non-IID data distribution
plot_federated_learning_results(non_iid_results, title_suffix='Non-IID')
```

```python
def plot_federated_learning_loss(results, title_suffix=''):
    import matplotlib.pyplot as plt
    import math
    
    data_distribution = results['data_distribution']
    client_loss_histories = results['client_loss_histories']
    
    num_clients = len(client_loss_histories)
    cols = 3  
    rows = math.ceil(num_clients / cols) + 1  
    
    # Create subplots with adjusted spacing
    fig, axes = plt.subplots(rows, cols, figsize=(cols * 5, rows * 4))
    axes = axes.flatten()  
    
    fig.suptitle(f'Federated Learning Loss Results ({data_distribution.upper()} Data)', fontsize=18, fontweight='bold', y=1.02)
    client_colors = plt.cm.tab10.colors  
    # Plot each client's loss history
    for idx, (client_id, loss_history) in enumerate(client_loss_histories.items()):
        ax = axes[idx]
        if loss_history:
            epochs = range(1, len(loss_history) + 1)
            losses = loss_history
            color = client_colors[idx % len(client_colors)]  
            ax.plot(epochs, losses, marker='o', color=color, linestyle='-', linewidth=2, markersize=6, label=f'Client {client_id}')
            ax.set_title(f'Client {client_id} Loss', fontsize=14, fontweight='medium')
            ax.set_xlabel('Epoch', fontsize=12)
            ax.set_ylabel('Loss', fontsize=12)
            ax.set_ylim(0, max(losses) + 0.1)  
            ax.grid(True, linestyle='--', alpha=0.7)
            ax.legend(loc='upper right', fontsize=10)
        else:
            ax.axis('off') 
    
   
    if num_clients < len(axes):
        axes[num_clients].axis('off')  
 
    for ax in axes[num_clients + 1:]:
        ax.axis('off')
    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    
    # Display the plots
    plt.show()

# Plot loss results for IID data distribution
plot_federated_learning_loss(iid_results, title_suffix='IID')

# Plot loss results for Non-IID data distribution
plot_federated_learning_loss(non_iid_results, title_suffix='Non-IID')
```

## Conclusion

In this simulation, we've demonstrated a simple **Federated Learning** which operates under both **IID** and **Non-IID** data distributions using the **Federated Averaging (FedAvg)** algorithm for model aggregation. The visualizations provide insights into individual client performances and the overall global model's progress.

### Key Takeaways:

- **Impact of Data Distribution:** Non-IID data distributions can pose significant challenges to FL, potentially leading to slower convergence or reduced global model accuracy due to issues like overfitting. Unlike IID distributions, where data across clients is uniform, Non-IID distributions introduce variability that can complicate the learning process.
  
- **FedAvg Aggregation:** Averaging client models remains an effective method for aggregating weights in FL, balancing individual client contributions. However, its effectiveness can be influenced by factors such as the degree of data heterogeneity and client participation rates.
